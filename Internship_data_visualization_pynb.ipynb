```python
plt.scatter(hr["EngagementSurvey"], hr["Salary"])

# Add labels and title
plt.xlabel("Engagement Survey Score")
plt.ylabel("Salary")
plt.title("Relationship Between Employee Engagement and Salary")
plt.show()
```

```python
recruitment_counts = hr["RecruitmentSource"].value_counts()

# Create a bar chart of recruitment source vs. number of employees hired
plt.bar(recruitment_counts.index, recruitment_counts.values)

# Add labels and title
plt.xlabel("Recruitment Source")
plt.ylabel("Number of Employees Hired")
plt.title("Number of Employees Hired by Recruitment Source")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.regplot(x="EngagementSurvey", y="Salary", data=hr)

# Add labels and title
plt.xlabel("Engagement Survey Score")
plt.ylabel("Salary")
plt.title("Relationship Between Employee Engagement and Salary")
plt.show()
```

```python
sns.violinplot(x="Department", y="Salary", data=hr)

# Add labels and title
plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Distribution of Salary by Department")
plt.xticks(rotation=90)

plt.show()
```

```python
plt.hist(airbnb["price"], bins=20)

# Add labels and title
plt.xlabel("Price (USD)")
plt.ylabel("Count")
plt.title("Distribution of Airbnb Listing Prices")
```

```python
sns.boxplot(x="room_type", y="number_of_reviews", data=airbnb)

# Add labels and title
plt.xlabel("Room Type")
plt.ylabel("Number of Reviews")
plt.title("Distribution of Airbnb Listing Reviews by Room Type")

plt.show()
```
