```python
def detectoutliers(data,col):
  movingmean=data[col].rolling(window=8,min_periods=6).mean().shift(1)
  movingstd=data[col].rolling(window=8,min_periods=6).mean().shift(1)


  upperlimit=movingmean+2*movingstd
  lowerlimit=movingmean-2*movingstd

  valuesabove_upperlmt=data[col].values>upperlimit
  valuesabove_lowerlmt=data[col].values<lowerlimit

  outliersboolean=np.logical_or(valuesabove_upperlmt,valuesabove_lowerlmt)
  outliers=np.where(outliersboolean)[0]

  outliers=data.index[outliers]

  print("before fixing outliers")
  fig=go.Figure()
  fig.add_trace(go.Scatter(x=data.index,y=data[col],name="close"))
  fig.add_trace(go.Scatter(x=data.index,y=upperlimit,name="upperlimit"))
  fig.add_trace(go.Scatter(x=data.index,y=lowerlimit,name="lowerlimit"))
  fig.show()

  print("numbers of the outliers detected -", len(outliers))


  data.loc[outliers,col]=np.nan
  data[col] =data[col].interpolate(method="linear")


  print("after fixing outliers" , len(outliers))


  fig=go.Figure()
  fig.add_trace(go.Scatter(x=data.index,y=data[col],name="close"))
  fig.add_trace(go.Scatter(x=data.index,y=upperlimit,name="upperlimit"))
  fig.add_trace(go.Scatter(x=data.index,y=lowerlimit,name="lowerlimit"))
  fig.show()


  return data[col]
```

```python
from statsmodels.tsa.stattools import kpss,adfuller 
```

```python
def statfind(data,col,method="Adfuller"):
  if method=="kpss":
    _, p, _ , _ = kpss(data[col], nlags="auto")
    if p > 0.05:
      return True
    else :
      return False 
  elif method == 'Adfuller':
    _, p, _, _, _, _ = adfuller(data[col])

    if p > 0.05: # we failed to reject the null hypothesis 
        return False
    else: # we rejected null hypothesis 
        return True

```

```python
answer=statfind(stock,"closelinear", method="kpss")
print("is stat ",answer)
```

```python
def smothingmovingavg(data:pd.Series,window_size=7):
  weights=np.arange(1,window_size+1)/window_size
  return data.rolling(window=window_size).apply(lambda x:np.sum(weights*x)/sum(weights))
```

```python
def expoweightsmovingavg(data:pd.Series,alpha=0.1):
  return data.ewm(alpha=alpha).mean()
```

```python
stock["movingavug"]=smothingmovingavg(stock["closelinear"])
stock["expomoving"]=expoweightsmovingavg(stock["closelinear"],alpha=0.4)
```

```python

def DoPredictions(Train, Test, real_time_mode = False): #   To simulate realtime scenario
    if real_time_mode: # pro active 
        # Inital Training 
        model = AutoReg(Train, lags = 4)
        #model = ARIMA(Train, order = (4, 0, 3))
        trained_model = model.fit()
        predictions = pd.Series()
        for i in range(0, len(Test), 7):

            # Get predictions for next 7 days 
            prediction_slice = trained_model.predict(start = len(Train) + i, end = len(Train) + i + 7)
            predictions = predictions.append(prediction_slice)

            # Retrain the model with those 7 days 
            newTrain = Train.append(Test.iloc[:i + 7])
            model = AutoReg(newTrain, lags = 4)
            #model = ARIMA(newTrain, order = (4, 0, 3))
            trained_model = model.fit()
        
        predictions = predictions.loc[Test.index] # making sure prediction and testing set have same indexes 
    else: # lazy
        model = AutoReg(Train, lags = 4)
        #model = ARIMA(Train, order = (4, 0, 3))
        trained_model = model.fit()
        predictions = trained_model.predict(start = len(Train), end = len(Train) + len(Test) + 1, dynamic = False)
        # print(predictions) # Returns Series object, with index as timestamps and values as predictions
    
    fig = go.Figure(data=go.Scatter(x=Test.index, y=Test.values, name = 'Actual'))
    fig.add_trace(go.Scatter(x=predictions.index, y=predictions.values, name =  'Predictions'))
    fig.show()
    
```
